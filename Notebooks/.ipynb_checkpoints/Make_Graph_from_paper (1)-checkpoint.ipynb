# GraphML from PDF

Converts a PDF, which is downloaded from a link, to a graph.

This is done using two functions:

1) A function to convert a PDF to Markdown
2) A step to generate a series (like a Markov chain) of question-answer pairs from the full text
3) A step to convert the question-answer pairs into sub-graphs, which are then assembled into a graph 

Uses ```marker```, install:

```bash
pip install marker-pdf -U
```

````python
import os
import json
import requests
import subprocess
from enum import Enum
from typing import Optional
from pydantic import BaseModel, Field
from urllib.parse import urlparse
from openai import OpenAI
from enum import Enum
from typing import Optional
from pydantic import BaseModel, Field
from tqdm.notebook import tqdm

# === CONFIG ===
PDF_URL = "https://www.nature.com/articles/s41467-025-57460-y.pdf"
PDF_DIR = "/home/mkychsu/pool/Saint Gobain"
MARKDOWN_DIR = "/home/mkychsu/pool/Saint Gobain/MD"
MODEL_URL = "http://localhost:1234/v1/"
MARKER_FORMAT = "markdown"
MODEL_NAME = "mistral"
max_context = 20000

topic='spongin'
````

### Download PDF, convert to Markdown, ask initial question

````python
def download_pdf(url, save_dir):
    os.makedirs(save_dir, exist_ok=True)
    filename = os.path.basename(urlparse(url).path)
    filepath = os.path.join(save_dir, filename)
    if not os.path.exists(filepath):
        r = requests.get(url)
        r.raise_for_status()
        with open(filepath, "wb") as f:
            f.write(r.content)
        print(f"Downloaded: {filepath}")
    else:
        print(f"Already downloaded: {filepath}")
    return filepath

def convert_pdf_with_marker(pdf_path, output_dir):
    base = os.path.splitext(os.path.basename(pdf_path))[0]
    
    os.makedirs(output_dir, exist_ok=True)
    md_file = os.path.join(output_dir, f"./{base}/{base}.md")
    
    if not os.path.exists(f"{output_dir}/{base}/{base}.md"):

        try:
            cmd = [
                "marker_single",
                pdf_path,
                "--output_dir", output_dir,
                "--output_format", MARKER_FORMAT
            ]
            subprocess.run(cmd, check=True)
            print(f"Converted PDF to Markdown using Marker, file: {md_file}")
        except:
            return None
        
    else:
        print (f"Using existing Markdown file: {md_file}")
        
    return md_file

def load_markdown(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read()
 
def answer_question(context_text, question, max_tokens=1024, temperature=0.2, top_p=0.1):
    client = OpenAI(api_key="none_needed", base_url=MODEL_URL)

    if context_text:
    
        query=(
                "Context information is below:\n"
                "---------------------\n"
                f"{context_text}\n"
                "---------------------\n"
                "Using the context above and your own knowledge, think step by step to respond to the task.\n"
                "\n*Instructions*\n"
                "Do not mention the context or information, just respond to the task based on the information provided. "
                "If appropriate, define all terms clearly, in a scientifically sound way.\n\n"
                f"Task: {question}\n\n"
                "Answer: "
                )
    else:
        query=(
                "Think step by step to answer the query.\n"
                "\n*Instructions*\n"
                "Do not mention the context or information, just respond to the task based on the information provided. "
                "If appropriate, define all terms clearly, in a scientifically sound way.\n\n"
                f"Task: {question}\n\n"
                "Answer: "
                )

    completion = client.beta.chat.completions.parse(
        model=MODEL_NAME,
        messages=[
            {
                "role": "system",
                "content": "You provide detailed answers from papers, with great detail and rationale.",
            },
            {
                "role": "user",
                "content": query,
            },
        ],
        max_tokens=max_tokens,
        temperature=temperature,
        top_p=top_p,
        
    )

    return completion.choices[0].message.content
   
pdf_path = download_pdf(PDF_URL, PDF_DIR)
md_path = convert_pdf_with_marker(pdf_path, MARKDOWN_DIR)
markdown_text = load_markdown(md_path)

question = f'What is {topic}?'

result = answer_question(markdown_text, question, max_tokens=1024, temperature=0.1) 

print(f"Question: {question}\n\nAnswer: {result}")
````

### Ask follow-up questions and construct list of q-a pairs

````python
i=0
N_q=64
qa_pairs=[]

for i in tqdm(range (N_q)):
    context_text=f'Question-Answer pair i={i}\n\nQuestion: {question}\n\nAnswer: {result}'
    question = answer_question(context_text=context_text, question=f'Make up a new, follow-up question that asks about a related topic that was not yet asked, but in the general topical area of {topic}. Do not answer it. Only provide a single question.', max_tokens=1024, temperature=0.3, top_p=0.1)
    print(f"***New question i={i}: {question}")
    result = answer_question(markdown_text, question, max_tokens=1024, temperature=0.1) 
    print(f"***Answer i={i}: {result}\n\n")

    qa_pairs.append (f'Question: {question}\n\nAnswer: {result}')
    i=i+1
````

### Convert list of q-a pairs into graph

````python
import networkx as nx
from pydantic import BaseModel
from typing import List
from openai import OpenAI
import json

class Node(BaseModel):
    id: str
    type: str
    

class Edge(BaseModel):
    source: str
    target: str
    relation: str

class KnowledgeGraph(BaseModel):
    nodes: List[Node]
    edges: List[Edge]

def update_graph_from_text(text_passage: str, G: nx.DiGraph):
    client = OpenAI(api_key="none_needed", base_url=MODEL_URL)

    completion = client.beta.chat.completions.parse(
        model=MODEL_NAME,
        messages=[
            {
                "role": "system",
                "content": (
                    "You are a scientific assistant extracting knowledge graphs from text. "
                    "Return a JSON with two fields: 'nodes' and 'edges'.\n"
                    "Each node must have 'id' and 'type'.\n"
                    "Each edge must have 'source', 'target', and 'relation'."
                ),
            },
            {
                "role": "user",
                "content": f"Text: {text_passage}\n\nExtract the knowledge graph in structured JSON.",
            },
        ],
        response_format=KnowledgeGraph,
        
        temperature=0.1,
        max_tokens=9000,
    )

    parsed = completion.choices[0].message.parsed#.model_dump()

    # Add nodes
    for node in parsed.nodes:
        G.add_node(node.id, type=node.type)

    # Add edges
    for edge in parsed.edges:
        G.add_edge(edge.source, edge.target, relation=edge.relation)

    # Print stats
    print(f"Graph updated; Nodes: {G.number_of_nodes()}, Edges: {G.number_of_edges()}")
````

````python
import networkx as nx

G = nx.DiGraph()

for passage in tqdm(qa_pairs):
    update_graph_from_text(passage, G)
````
